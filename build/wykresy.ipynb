```python
import numpy as np
import matplotlib.pyplot as plt
import sys

# Read data from .plt file
data = np.genfromtxt('UVP.plt', skip_header=3)
np.set_printoptions(threshold=sys.maxsize)
print(data)
```

```python
# Extract variables from data
x = data[:, 0]
y = data[:, 1]
u = data[:, 2]
v = data[:, 3]
p = data[:, 4]

# Reshape variables into 2D grids
nx = len(np.unique(x))
ny = len(np.unique(y))
x_grid = x.reshape((ny, nx))
y_grid = y.reshape((ny, nx))
u_grid = u.reshape((ny, nx))
v_grid = v.reshape((ny, nx))
p_grid = p.reshape((ny, nx))

# Create contour plots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))

# Contour plot for u
c1 = ax1.contourf(x_grid, y_grid, u_grid, cmap='jet')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Contour plot of u')

# Contour plot for v
c2 = ax2.contourf(x_grid, y_grid, v_grid, cmap='jet')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Contour plot of v')

# Contour plot for p
c3 = ax3.contourf(x_grid, y_grid, p_grid, cmap='jet')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('Contour plot of p')

# Add colorbars
cbar1 = fig.colorbar(c1, ax=ax1)
cbar1.set_label('u')
cbar2 = fig.colorbar(c2, ax=ax2)
cbar2.set_label('v')
cbar3 = fig.colorbar(c3, ax=ax3)
cbar3.set_label('p')

# Display the plots
plt.tight_layout()
plt.show()
```
